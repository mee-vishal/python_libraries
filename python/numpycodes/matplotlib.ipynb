```python
# labels and titles
import matplotlib.pyplot as plt
import numpy as np
x=np.array([3,4,1,6,9,2,3])
y=np.array([1,7,3,8,4,2])

plt.plot(x,color='r',linewidth='1.5')
plt.plot(y,color='y',linewidth='5.5')
plt.xlabel("avg oxygen")
plt.ylabel("our calorie")
plt.title("my body")
plt.show()
```

```python
# we can also change the font property of title and labels
x=np.array([3,4,1,6,9,2,3])
y=np.array([1,7,3,8,4,2])
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'green','size':20}
plt.plot(x,color='r',linewidth='1.5')
plt.plot(y,color='y',linewidth='5.5')
plt.xlabel("avg oxygen",fontdict=font1)
plt.ylabel("our calorie",fontdict=font1)
plt.title("my body",fontdict=font2)
plt.show()
```

```python
# we can also set positions
# loc
x=np.array([3,4,1,6,9,2,3])
y=np.array([1,7,3,8,4,2])
font1={'family':'serif','color':'blue','size':20}
font2={'family':'serif','color':'green','size':20}
plt.plot(x,color='r',linewidth='1.5')
plt.plot(y,color='y',linewidth='5.5')
plt.xlabel("avg oxygen",fontdict=font1,loc='right')
plt.ylabel("our calorie",fontdict=font1)
plt.title("my body",fontdict=font2,loc='right')
plt.show()
```

```python
# display the multiple plots with subplot ->>we can draw multiple plots 
import matplotlib.pyplot as plt
import numpy as np


# plot1
x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
plt.subplot(2,1,1)  
plt.plot(x,y)
# the figure has 1-row 2-columns and this plot is the 1st plot

# plot2
x=np.array([1,1,2,3])
y=np.array([1,9,7,2])
plt.subplot(2,1,2) 
plt.plot(x,y)  
# the figure has 1-row 2-columns and this plot is the 2nd plot
plt.show()
```

```python
# plt.subplot(2,1,2) it means there will be 2 rows 1 column and this will be the 2nd element of that column
```

```python
# angles 
#  the default angle is at the x axis but we can change the start angle by specifying start angle parameter
import matplotlib.pyplot as plt
import numpy as np
x=np.array([5,3,2,1])
label=['apple','banana','mango','cherry']
plt.pie(x,labels=label,startangle=90)
plt.show()
```
